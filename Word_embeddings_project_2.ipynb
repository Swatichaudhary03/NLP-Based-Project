Import Required Libraries

```python
import pandas as pd
import nltk
import re
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

!pip install gensim
from gensim.models import Word2Vec
```

```python
df = pd.read_csv("Reviews.csv")
df
```

```python
df.head()
```

Text Cleaning Function

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

df['Text'].fillna('', inplace=True) # Fill NaN values with empty string
df['cleaned_review'] = df['Text'].apply(clean_text)
```

Tokenization

```python
df['tokens'] = df['cleaned_review'].apply(word_tokenize)
df[['cleaned_review', 'tokens']].head()
```

```python
df['tokens'] = df['cleaned_review'].apply(word_tokenize)
df[['cleaned_review', 'tokens']].head()
```

Stopword Removal

```python
stop_words = set(stopwords.words('english'))

df['tokens_no_stopwords'] = df['tokens'].apply(
    lambda x: [word for word in x if word not in stop_words]
)
```

Stemming

```python
stemmer = PorterStemmer()

df['stemmed_tokens'] = df['tokens_no_stopwords'].apply(
    lambda x: [stemmer.stem(word) for word in x]
)
```

Lemmatization

```python
lemmatizer = WordNetLemmatizer()

df['lemmatized_tokens'] = df['tokens_no_stopwords'].apply(
    lambda x: [lemmatizer.lemmatize(word) for word in x]
)
```

Final Preprocessed Text

```python
df['final_text'] = df['lemmatized_tokens'].apply(lambda x: " ".join(x))
df[['Text', 'final_text']].head()
```

TEXT REPRESENTATION MODELS

## Bag-of-Words (BoW) Representation

```python
count_vectorizer = CountVectorizer()
bow_matrix = count_vectorizer.fit_transform(df['final_text'])

print("Bag-of-Words matrix shape:", bow_matrix.shape)
print("First 5 feature names:", count_vectorizer.get_feature_names_out()[:5])
```

TF-IDF (Term Frequencyâ€“Inverse Document Frequency)

```python
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df['final_text'])

print(X_tfidf.toarray()[:2])
```

Word Embeddings â€“ Word2Vec

```python
sentences = df['lemmatized_tokens'].tolist()

word2vec_model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

word2vec_model.wv['good']
```
